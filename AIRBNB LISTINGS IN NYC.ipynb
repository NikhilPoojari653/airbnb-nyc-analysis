## 2: Data Cleaning

```python
# Drop irrelevant columns
columns_to_drop = [
    'id', 'listing_url', 'scrape_id', 'last_scraped', 'source', 'name', 
    'description', 'neighborhood_overview', 'picture_url', 'host_url', 
    'host_name', 'host_about', 'host_thumbnail_url', 'host_picture_url', 
    'host_verifications', 'calendar_updated', 'calendar_last_scraped',
    'first_review', 'last_review', 'license'
]
df.drop(columns=columns_to_drop, inplace=True)
```

```python
# Convert price to numeric (remove $ and commas)
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Drop rows where price is 0 or extremely high (outliers)
df = df[df['price'] > 0]
df = df[df['price'] < 1000]  # Cap at 1000 USD for better visuals
```

```python
# Check missing values
missing_values = df.isnull().sum().sort_values(ascending=False)
```

```python
# Display cleaned dataset shape and top missing columns
df.shape, missing_values.head(10)
```

##  3: EDA

```python
#A. Listings by Borough (Plot)
plt.figure(figsize=(10, 6))
sns.countplot(
    data=df, 
    x='neighbourhood_group_cleansed', 
    order=df['neighbourhood_group_cleansed'].value_counts().index,
    palette='Set2'
)
plt.title('Number of Listings by Borough')
plt.xlabel('Borough')
plt.ylabel('Count of Listings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig()
```

```python
# 3.2 Room‑type distribution vs borough
plt.figure(figsize=(10, 6))
sns.countplot(
    data=df,
    x="neighbourhood_group_cleansed",
    hue="room_type",
    order=df["neighbourhood_group_cleansed"].value_counts().index,
)
plt.title("Room‑Type Breakdown per Borough")
plt.xlabel("Borough")
plt.ylabel("Listings Count")
plt.legend(title="Room Type")
savefig("roomtype_by_borough")
plt.show()
```

```python
# 3.3 Price distribution overall & by borough
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df["price"], bins=50, kde=True, ax=ax[0])
ax[0].set_title("Price Distribution (USD)")
ax[0].set_xlabel("Nightly Price")

sns.boxplot(
    data=df,
    x="neighbourhood_group_cleansed",
    y="price",
    order=df["neighbourhood_group_cleansed"].value_counts().index,
    ax=ax[1],
)
ax[1].set_title("Price by Borough (Boxplot)")
ax[1].set_xlabel("Borough")
ax[1].set_ylabel("Nightly Price")
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=30)

savefig("price_distribution")
plt.show()
```

```python
# 3.4 Average price by neighbourhood bar chart
price_neigh = (
    df.groupby("neighbourhood")["price"].mean().sort_values(ascending=False).head(20)
)
plt.figure(figsize=(10, 6))
price_neigh.plot(kind="bar")
plt.title("Top 20 Most Expensive NYC Neighbourhoods (Average Price)")
plt.ylabel("Average Nightly Price (USD)")
plt.xlabel("Neighbourhood")
plt.xticks(rotation=75)
plt.show()
```

```python
# 3.5 Availability analysis
plt.figure(figsize=(8, 5))
sns.histplot(df["availability_365"], bins=50, color="tab:blue")
plt.title("Distribution of Availability (Days per Year)")
plt.xlabel("Days Available")
plist = [0, 30, 90, 180, 365]
plt.xticks(plist)
plt.show()
```

```python
# Availability vs price scatter (sample 5k for performance)
sample = df.sample(5000, random_state=42)
plt.figure(figsize=(8, 5))
plt.scatter(sample["availability_365"], sample["price"], alpha=0.3)
plt.title("Price vs Availability (Sample of 5k Listings)")
plt.xlabel("Days Available per Year")
plt.ylabel("Nightly Price (USD)")
plt.show()
```

```python
# 3.6 Host behaviour – multi‑listing hosts
host_counts = df["host_id"].value_counts()

multi_hosts = host_counts[host_counts > 1]
print("\nHosts with >1 listing:", multi_hosts.count(), "(of", host_counts.count(), ")")

top_hosts = multi_hosts.head(20)
plt.figure(figsize=(10, 6))
(top_hosts.sort_values(ascending=True)  # horizontal bar
          .plot(kind="barh"))
plt.title("Top 20 Hosts by Number of Listings")
plt.xlabel("Number of Listings")
plt.ylabel("Host ID")
plt.show()
```

```python
# 3.7 Reviews & ratings
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df, x="review_scores_rating", y="price", alpha=0.3, hue="room_type", legend=False
)
plt.title("Price vs Review Rating")
plt.xlabel("Review Score (0–100)")
plt.ylabel("Nightly Price (USD)")
plt.show()
```

```python
# Reviews per month by borough
reviews_borough = (
    df.groupby("neighbourhood_group_cleansed")["reviews_per_month"].mean().sort_values()
)
plt.figure(figsize=(8, 5))
reviews_borough.plot(kind="bar", color="tab:green")
plt.title("Average Reviews per Month by Borough")
plt.ylabel("Reviews per Month (Avg)")
plt.xlabel("Borough")
plt.xticks(rotation=30)
plt.show()
```

```python
# 4.Clustering – quick host segmentation using K‑Means
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Feature set: price, accommodates, availability_365, review_scores_rating
features = df[["price", "accommodates", "availability_365", "review_scores_rating"]].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

k = 4  # arbitrary for demonstration
kmeans = KMeans(n_clusters=k, random_state=42)
features["cluster"] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=features, x="price", y="review_scores_rating", hue="cluster", palette="Set1", alpha=0.5
)
plt.title("K‑Means Clusters: Price vs Rating")
plt.xlabel("Price (USD)")
plt.ylabel("Review Score")
plt.legend(title="Cluster")
plt.show()
```

## 5. Conclusions – key business insights
### Key Insights:
### 1. Manhattan and Brooklyn dominate the listing count, reflecting higher tourism activity.
### 2. Entire apartments command the highest median price, especially in Manhattan.
### 3. Price hotspots concentrate in mid‑town Manhattan; outer boroughs are generally cheaper.
### 4. Listings with near‑full‑year availability likely indicate professional hosts; they can be targeted for regulation.
### 5. Host segmentation shows distinct clusters (e.g., budget high‑availability vs premium low‑availability). These clusters guide differentiated marketing strategies.
### 6. No clear monotonic relationship between review scores and price — leverage unique experiences to justify premium.
